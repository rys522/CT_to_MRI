```python
import glob
from pathlib import Path

import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
log_v = 'log_2025_02_07_conditional_test'
run_idx = 0
log_root = Path(f"../log/{log_v}/{run_idx:05d}_test/test")
res_files = glob.glob(str(log_root) + '/*.mat')
print(len(res_files))
```

```python
vmax, vmin = 1, 0
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(wspace=0, hspace=0, left=0, right=1, top=1, bottom=0)
fig.patch.set_facecolor('black')
for idx in range(10):
    res_mat = loadmat(res_files[idx])
    cond = res_mat['cond'].squeeze()
    out = res_mat['out'].squeeze()
    img = res_mat['img'].squeeze()

    plt.subplot(3,10,idx+1); plt.imshow(cond, cmap='gray', vmax=vmax, vmin=vmin); plt.axis('off')
    plt.subplot(3,10,idx+11); plt.imshow(out, cmap='gray', vmax=None, vmin=vmin); plt.axis('off')
    plt.subplot(3,10,idx+21); plt.imshow(img, cmap='gray', vmax=vmax, vmin=vmin); plt.axis('off')


plt.tight_layout()
plt.subplots_adjust(left=0, right=1, top=1, bottom=0, wspace=0, hspace=0)
plt.show()
```
