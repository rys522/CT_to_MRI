```python
import glob
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
```

```python
log_v = 'log_2025_02_07_unconditional'
run_idx = 0 
run_ep = 98
log_root = Path(f"../log/{log_v}/{run_idx:05d}_train/test/ep_{run_ep}")
res_files = glob.glob(str(log_root) + '/*.mat')
print(len(res_files))
```

```python
vmax, vmin = 1, 0
fig = plt.figure(figsize=(10, 5))
plt.subplots_adjust(wspace=0, hspace=0, left=0, right=1, top=1, bottom=0)
fig.patch.set_facecolor('black')
for idx in range(30):
    res_mat = loadmat(res_files[idx])
    out = np.array(res_mat['out']).squeeze()

    plt.subplot(3,10,idx+1); plt.imshow(out, cmap='gray', vmax=vmax, vmin=vmin); plt.axis('off')


plt.tight_layout()
plt.subplots_adjust(left=0, right=1, top=1, bottom=0, wspace=0, hspace=0)
plt.show()
```
